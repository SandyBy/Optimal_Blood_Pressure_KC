```python
df_group = df.groupby("icustay_id")
```

```python
def calc_percent(df, timerange=[0, 1440], col='sysbp', valuerange=[115, 125]):
    start = timerange[0]
    end = timerange[1]
    temp = df[(df['chartoffset']>=start) & (df['chartoffset']<=end)]
    non_na = temp[temp[col].notnull()]
    num_nonna = non_na.shape[0]
    if num_nonna==0:
        return np.nan
    value_min, value_max = valuerange[0], valuerange[1]
    range_num = non_na[(non_na[col]>=value_min) & (non_na[col]<=value_max)].shape[0]
    return range_num / num_nonna
```

```python
result = df_group.apply(lambda x: calc_percent(x, timerange=[0, 1440], col='meanbp', valuerange=[106.0754, 355]))
```

```python
result_df = pd.DataFrame(result).reset_index()
result_df.columns = ["icustay_id", "mbp_24_inrange_AKI"]
```
